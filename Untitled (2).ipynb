```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread("salah.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
filter_size = (5, 5)
average_filter_kernel = np.ones(filter_size, np.float32) / (filter_size[0] * filter_size[1])
average_filtered_image = cv2.filter2D(image_rgb, -1, average_filter_kernel)
gaussian_filtered_image = cv2.GaussianBlur(image_rgb, filter_size, sigmaX=0)
plt.figure(figsize=(14, 6))
plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(average_filtered_image)
plt.title("Average Filter (5x5)")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(gaussian_filtered_image)
plt.title("Gaussian Filter (5x5)")
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
noisy_image = cv2.imread("salah.jpg", cv2.IMREAD_GRAYSCALE)
median_filtered = cv2.medianBlur(noisy_image, 5)
min_filtered = cv2.erode(noisy_image, np.ones((5, 5), np.uint8))
max_filtered = cv2.dilate(noisy_image, np.ones((5, 5), np.uint8))
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.imshow(noisy_image, cmap="gray")
plt.title("Original Image (Salt-and-Pepper Noise)")
plt.axis("off")
plt.subplot(2, 2, 2)
plt.imshow(median_filtered, cmap="gray")
plt.title("Rank Filter (Rank = 13) - Median Filter")
plt.axis("off")
plt.subplot(2, 2, 3)
plt.imshow(min_filtered, cmap="gray")
plt.title("Rank Filter (Rank = 1) - Min Filter")
plt.axis("off")
plt.subplot(2, 2, 4)
plt.imshow(max_filtered, cmap="gray")
plt.title("Rank Filter (Rank = 25) - Max Filter")
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def mylaplacian(image):
    image = image.astype(np.float32)
    height, width = image.shape
    laplacian_image = np.zeros_like(image, dtype=np.float32)
    for x in range(1, height - 1):
        for y in range(1, width - 1):
            horizontal = image[x, y + 1] + image[x, y - 1] - 2 * image[x, y]
            vertical = image[x + 1, y] + image[x - 1, y] - 2 * image[x, y]
            laplacian_image[x, y] = horizontal + vertical
    return np.clip(laplacian_image, 0, 255).astype(np.uint8)
image = cv2.imread("Picture1.jpg", cv2.IMREAD_GRAYSCALE)
manual_laplacian = mylaplacian(image)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)
plt.figure(figsize=(15, 8))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(manual_laplacian, cmap="gray")
plt.title("Manual Laplacian (mylaplacian)")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(sobel_combined, cmap="gray")
plt.title("Sobel Filter (Combined)")
plt.axis("off")
plt.tight_layout()
plt.show()
```
